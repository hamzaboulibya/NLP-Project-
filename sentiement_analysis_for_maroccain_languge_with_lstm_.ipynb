```python
import re

def pre_processing(df, field):
    lis=[]
    # Replace URLs, user mentions, and hashtags
    df[field] = df[field].replace(r'http\S+', 'URL', regex=True)
    df[field] = df[field].replace(r'www\S+', 'URL', regex=True)
    df[field] = df[field].replace(r'@[^\s]+', 'USER', regex=True)
    df[field] = df[field].replace(r'#[^\s]+', 'HASHTAG', regex=True)
    df = df[df[field].apply(lambda x: len(re.findall(r'[\u0600-\u06FF]+', x))) > 1]
    # df[field] = df[field].apply(strip_tatweel)
    # df[field] = df[field].apply(strip_tashkeel)

    symbols1 = re.compile('[/(){}\[\]\|,;]')
    symbols2 = re.compile("[@|(|)|\||:|>|<|_|#|\.|+|÷|×|'|!|\?|٪|؟\|&;*|[{|}]|\-|،|_’;!:^&%//]")
    arabic_punctuations = '''`÷×؛<>_()*&^%][ـ،/:"؟.,'{}~¦+|!”…“–ـ'''

    df[field] = df[field].apply(lambda x: re.sub(symbols1, ' ', x))
    df[field] = df[field].apply(lambda x: re.sub(symbols2, ' ', x))
    df[field] = df[field].apply(lambda x: re.sub('\n', ' ', x))
    # df[field]  =strip_tashkeel(df[field])
    # df[field]=strip_diacritics(df[field])
    # df[field]=normalize_ligature(df[field])
    for i in df['comment']:
      lis.append(i)


    return df

df = pre_processing(df, 'comment')
```

### Embedding Representation

```python
sent_length=100
embedded_docs=pad_sequences(onehot_repr,padding='post',maxlen=sent_length)
print(embedded_docs)
```

```python
embedded_docs[1]
```

```python
## Creating model
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
import numpy as np
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

### Model Training

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
from tensorflow.keras.layers import Dropout
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

### Performance










```python
y_pred=model.predict(X_test)
```
